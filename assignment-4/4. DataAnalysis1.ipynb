```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("C:/Users/Ritesh/OneDrive/Desktop/HousingData.csv")
df
```

```python
df.columns
```

```python
x = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
x
```

```python
y=df["MEDV"]
```

```python
y
```

```python
#Model Training: main data set split training: 75% and testing 25%
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
model = LinearRegression()
x_train.replace(np.nan,x_train.min(),inplace=True)
x_test.replace(np.nan,x_test.min(),inplace=True)
x_train.isnull().sum()
x_test.isnull().sum()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```
